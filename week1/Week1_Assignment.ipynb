### **Step 1: Load the Dataset**

```python
import pandas as pd

# 1. Load the CSV dataset into a Pandas DataFrame
file_name = 'Combined_dataset.csv'
df = pd.read_csv(file_name)

print(f'Successfully loaded {file_name}')
```

### **Step 2: Explore data**

```python
# Display the first 5 rows
print('First 5 rows:')
display(df.head())
```

```python
# Display the last 5 rows
print('\nLast 5 rows:')
display(df.tail())
```

```python
# Get the shape of the DataFrame (rows, columns)
print(f'\nDataFrame shape: {df.shape}')
```

```python
# List all columns and their data types
print('\nColumn information and data types:')
df.info()
```

### **Step 3: Handle missing values**

```python
import re

# Identify missing values
print('Missing values before handling:')
display(df.isnull().sum())
```

```python
# Handle missing values for specific columns
# Fill object columns with 'Unknown'
for col in ['what_customers_said', 'seller_name', 'seller_information', 'variations', 'videos']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
```

```python
# Fill 'discount' (numeric) with 0 (or mean/median if preferred)
if 'discount' in df.columns:
    df['discount'] = df['discount'].fillna(0)
```

```python
# Convert 'final_price' to numeric
if 'final_price' in df.columns:
    # Remove all characters that are not digits or a decimal point
    df['final_price'] = df['final_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
    # Convert to numeric, coercing errors to NaN
    df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
    # Fill NaNs with 0 (for any values that couldn't be converted)
    df['final_price'] = df['final_price'].fillna(0)

print('\nMissing values after handling and data type conversions:')
display(df.isnull().sum())
print('\nDataFrame info after handling missing values and converting types:')
df.info()
```

### **Step 4: Perform basic operations**

```python
# 1) Filter rows where 'ratings_count' (proxy for Quantity) is greater than 0
# Since 'Quantity' column is not available, 'ratings_count' is used as a proxy for user engagement.
if 'ratings_count' in df.columns:
    df_filtered = df[df['ratings_count'] > 0]
    print(f'\nShape after filtering for ratings_count > 0: {df_filtered.shape}')
    print('First 5 rows of filtered data:')
    display(df_filtered.head())
else:
    df_filtered = df.copy()
    print("No 'ratings_count' column found for filtering. Keeping all rows.")

# 2) Select relevant columns for the cleaned dataset
selected_columns = ['product_id', 'title', 'product_description', 'final_price', 'ratings_count']
existing_selected_columns = [col for col in selected_columns if col in df_filtered.columns]

df_selected = df_filtered[existing_selected_columns]
print(f'\nShape after selecting columns: {df_selected.shape}')
print('First 5 rows of selected columns:')
display(df_selected.head())
```

### **Step 5. Remove duplicates**

```python
# Re-deriving df_filtered and df_selected to ensure consistency, in case of previous state issues.

# Filter rows where 'ratings_count' (proxy for Quantity) is greater than 0
if 'ratings_count' in df.columns:
    df_filtered = df[df['ratings_count'] > 0]
else:
    df_filtered = df.copy()
    print("No 'ratings_count' column found for filtering. Keeping all rows.")

# Select relevant columns for the cleaned dataset
selected_columns = ['product_id', 'title', 'product_description', 'final_price', 'ratings_count']
existing_selected_columns = [col for col in selected_columns if col in df_filtered.columns]
df_selected_fresh = df_filtered[existing_selected_columns]

print(f'Shape of df_selected before duplicate removal: {df_selected_fresh.shape}')
print('First 5 rows of df_selected before duplicate removal:')
display(df_selected_fresh.head())

initial_rows_dup = df_selected_fresh.shape[0]
df_cleaned = df_selected_fresh.drop_duplicates()
num_duplicates_removed = initial_rows_dup - df_cleaned.shape[0]

print(f'\nNumber of duplicate rows removed: {num_duplicates_removed}')
print(f'Shape after removing duplicates: {df_cleaned.shape}')
print('First 5 rows after removing duplicates:')
display(df_cleaned.head())
```

### **Step 6: Create a derived column (total_amount = price * quantity)**

```python
if 'final_price' in df_cleaned.columns and 'ratings_count' in df_cleaned.columns:
    df_cleaned['total_amount'] = df_cleaned['final_price'] * df_cleaned['ratings_count']
    print('\nDataFrame with new \'total_amount\' column:')
    display(df_cleaned.head())
else:
    print("Cannot create 'total_amount' column: 'final_price' or 'ratings_count' columns are missing or not numeric.")
```

### **Step 7: Save the cleaned dataset as a new CSV file**

```python
output_file_path = 'Combined_dataset_cleaned.csv'
df_cleaned.to_csv(output_file_path, index=False)
print(f'Cleaned data saved to: {output_file_path}')
```

```python
print(f'\nShape of cleaned dataset: {df_cleaned.shape}\n')
print(f'Number of rows: {df_cleaned.shape[0]}')
print(f'Number of columns: {df_cleaned.shape[1]}')

print('\nInfo of the cleaned dataset:')
df_cleaned.info()

print('\nFirst 5 rows of the cleaned dataset:')
display(df_cleaned.head())
```
